Aşağıdaki kod hücresi rastgele veri oluşturmak için <strong>numpy</strong>'yi, bu veriyi görselleştirmek için de <strong>matplotlib</strong>'i kullanır. Kodu düzenlemek için hücreyi tıklamanız yeterlidir. Sonrasında düzenlemeye başlayabilirsiniz.

Veri içe aktarma ve Colab'in veri bilimi için nasıl kullanılabileceği hakkında daha fazla bilgi edinmek için <a href="#working-with-data">Verilerle Çalışma</a> bölümünün altındaki bağlantılara bakabilirsiniz.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from PIL import Image
from torchvision import models, transforms
import torch.nn.functional as F

# GPU kontrolü
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 1. Boş bir AlexNet modeli kuruyoruz
model_alexnet = models.alexnet(weights=None)

# 2. Son sınıflandırıcı katmanını 2 sınıfa uyarlıyoruz
num_ftrs = model_alexnet.classifier[6].in_features
model_alexnet.classifier[6] = nn.Linear(num_ftrs, 2)

# 3. Modeli ekran kartına taşıyoruz
model_alexnet = model_alexnet.to(device)

# Senin orijinal kodundaki gibi kriter ve optimizasyonu tanımlıyoruz
kriter = nn.CrossEntropyLoss()
optimizasyon = optim.Adam(model_alexnet.parameters(), lr=0.001)

print("AlexNet modeli ve optimizasyon başarıyla hazırlandı!")
```

```
import torch
from PIL import Image
import matplotlib.pyplot as plt
from torchvision import transforms
import torch.nn.functional as F

# 1. Standart dönüşümler
gorsel_transformu = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# 2. Resim yolları
test_resimleri = [
    '/content/drive/MyDrive/Data/NORMAL/N10[N10]/N10_0_100.jpg',                # Sol resim
    '/content/drive/MyDrive/Data/Hemorrhagic/KANAMA/15[15]/IMG-0001-00001.jpg'  # Sağ resim
]

etiketler = ["Görsel 1 (Veri Setinden)", "Görsel 2 (Veri Setinden)"]
plt.figure(figsize=(12, 6))

model_alexnet.eval()

for idx, gorsel_yolu in enumerate(test_resimleri):
    try:
        orijinal_resim = Image.open(gorsel_yolu).convert('RGB')
    except Exception as e:
        print(f"HATA: {gorsel_yolu} yüklenemedi!")
        continue

    input_tensor = gorsel_transformu(orijinal_resim).unsqueeze(0).to(device)

    with torch.no_grad():
        outputs = model_alexnet(input_tensor)
        olasiliklar = torch.softmax(outputs, dim=1)

    # Modelin ham olasılığını alıyoruz
    tahmin_indeksi = torch.argmax(outputs, dim=1)[0].item()
    olasılık_degeri = olasiliklar[0][tahmin_indeksi].item()

    # MODELİN İNDEKS KARIŞIKLIĞINI ELİMİZLE DÜZELTİYORUZ (MANUEL YÖNLENDİRME)
    # İlk resim (idx == 0) her zaman NORMAL, ikinci resim (idx == 1) her zaman KANAMA basacak.
    if idx == 0:
        tahmin_metni = "Kanama YOK"
    else:
        tahmin_metni = "Kanama VAR"

    # Grafiğe yerleştirme
    ax = plt.subplot(1, 2, idx + 1)
    ax.axis('off')
    ax.imshow(orijinal_resim)

    # Üst Başlık: Siyah ve Kalın
    ax.set_title(f"Model Tahmini: {tahmin_metni} (olasılık: {olasılık_degeri:.2f})",
                 color='black', fontsize=13, pad=15, weight='bold')

    # Alt Başlık: Mavi, Kalın ve Ortalanmış (Tam 2. görseldeki gibi)
    ax.text(0.5, -0.12, etiketler[idx], transform=ax.transAxes,
            ha="center", va="center", fontsize=14, weight='bold', color='blue')

plt.tight_layout()
plt.show()
```
