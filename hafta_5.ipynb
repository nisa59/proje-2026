```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread('beyin.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


Sx = 1.5
Sy = 1.2


T_matrisi = np.float32([[Sx, 0, 0], 
                        [0, Sy, 0]])


yukseklik, genislik = img.shape[:2]
yeni_genislik = int(genislik * Sx)
yeni_yukseklik = int(yukseklik * Sy)


olceklenmis_img = cv2.warpAffine(img_rgb, T_matrisi, (yeni_genislik, yeni_yukseklik))


plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title(f"Orijinal Görüntü {genislik}x{yukseklik}")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title(f"Ölçeklenmiş (Sx={Sx}, Sy={Sy}) {yeni_genislik}x{yeni_yukseklik}")
plt.imshow(olceklenmis_img)
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread('beyin.jpg')
if img is None:
    print("Görsel bulunamadı! Lütfen dosya adını kontrol edin.")
else:
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    rows, cols = img_rgb.shape[:2]

    angle = 30 
    theta = np.radians(angle)

  
    cos_t = np.cos(theta)
    sin_t = np.sin(theta)
    
   
    M = np.float32([[cos_t, -sin_t, 0],
                    [sin_t,  cos_t, 0]])

    
  
    rotated_img = cv2.warpAffine(img_rgb, M, (cols, rows))


    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.title("Orijinal Görüntü")
    plt.imshow(img_rgb)
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title(f"{angle} Derece Döndürülmüş")
    plt.imshow(rotated_img)
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread('beyin.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


tx = 100 
ty = 50


M = np.float32([[1, 0, tx], 
                [0, 1, ty]])


yukseklik, genislik = img.shape[:2]


otelenmis_img = cv2.warpAffine(img_rgb, M, (genislik, yukseklik))


plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Orijinal Görüntü (f(x,y))")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title(f"Ötelenmiş Görüntü (tx={tx}, ty={ty})")
plt.imshow(otelenmis_img)
plt.axis('off')

plt.tight_layout()
plt.show()
```
