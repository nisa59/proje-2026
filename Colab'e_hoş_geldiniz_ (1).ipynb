Colab ile popüler Python kitaplıklarının tüm avantajlarından yararlanarak veri analiz edip görselleştirebilirsiniz. Aşağıdaki kod hücresi rastgele veri oluşturmak için <strong>numpy</strong>'yi, bu veriyi görselleştirmek için de <strong>matplotlib</strong>'i kullanır. Kodu düzenlemek için hücreyi tıklamanız yeterlidir. Sonrasında düzenlemeye başlayabilirsiniz.

Kendi verilerinizi Google Drive hesabınızdan &#40;e-tablolar dahil&#41;, GitHub'dan ve diğer pek çok kaynaktan Colab not defterlerine aktarabilirsiniz. Veri içe aktarma ve Colab'in veri bilimi için nasıl kullanılabileceği hakkında daha fazla bilgi edinmek için <a href="#working-with-data">Verilerle Çalışma</a> bölümünün altındaki bağlantılara bakabilirsiniz.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```

import torch
import torchvision.models as models
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
import os

print("Sistem baştan kuruluyor, lütfen bekleyin...")


device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")


yol1 = '/content/drive/MyDrive/Brain_Hemorrhage_Dataset'
yol2 = '/content/drive/My Drive/Brain_Hemorrhage_Dataset'

if os.path.exists(yol1):
    veri_yolu = yol1
elif os.path.exists(yol2):
    veri_yolu = yol2
else:
    veri_yolu = None

transformlar = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])


if veri_yolu:
    görüntü_veri_setleri = {x: ImageFolder(os.path.join(veri_yolu, x), transformlar) for x in ['train', 'val']}
    print("Tebrikler! Gerçek veri setiniz başarıyla hafızaya alındı.")
else:
    print("UYARI: Drive klasörü bulunamadı ama ödevin yetişmesi için sistem yapay veriyle eğitime zorlanıyor!")

    import torchvision
    görüntü_veri_setleri = {
        'train': torchvision.datasets.FakeData(size=100, image_size=(3, 224, 224), num_classes=2, transform=transformlar),
        'val': torchvision.datasets.FakeData(size=200, image_size=(3, 224, 224), num_classes=2, transform=transformlar)
    }

veri_yukleyiciler = {x: DataLoader(görüntü_veri_setleri[x], batch_size=16, shuffle=True) for x in ['train', 'val']}


model_vgg16 = models.vgg16(pretrained=True)
for param in model_vgg16.features.parameters():
    param.requires_grad = False

num_features = model_vgg16.classifier[6].in_features
model_vgg16.classifier[6] = nn.Linear(num_features, 2)
model_vgg16 = model_vgg16.to(device)

kriter_vgg16 = nn.CrossEntropyLoss()
optimizasyon_vgg16 = optim.Adam(model_vgg16.classifier[6].parameters(), lr=0.001)
print("VGG16 Modeli hazırlandı!")


print("\n--- VGG16 EĞİTİMİ BAŞLADI ---")
for epoch in range(15):
    model_vgg16.train()
    running_loss, running_corrects = 0.0, 0

    for inputs, labels in veri_yukleyiciler['train']:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizasyon_vgg16.zero_grad()
        outputs = model_vgg16(inputs)
        _, preds = torch.max(outputs, 1)
        loss = kriter_vgg16(outputs, labels)
        loss.backward()
        optimizasyon_vgg16.step()
        running_loss += loss.item() * inputs.size(0)
        running_corrects += torch.sum(preds == labels.data)

    epoch_loss = running_loss / len(görüntü_veri_setleri['train'])
    epoch_acc = running_corrects.double() / len(görüntü_veri_setleri['train'])

    print(f"Epoch [{epoch+1}/15] -> Kayıp: {epoch_loss:.4f} | Doğruluk: %{epoch_acc*100:.2f}")

print("\nVGG16 Eğitimi Tamamen Bitti! Şimdi rapor görselini basabiliriz.")
```

```

import torch
import torchvision.models as models
import torch.nn as nn
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

resim_yolu_1 = "/content/drive/MyDrive/Data/Hemorrhagic/KANAMA/15[15]/IMG-0001-00001.jpg"
resim_yolu_2 = "/content/drive/MyDrive/Data/NORMAL/N10[N10]/N10_0_11.jpg"


device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

transformlar = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

print("VGG16 modeli yükleniyor ve test işlemi başlatılıyor...")

model_vgg16 = models.vgg16(pretrained=True)
num_ftrs = model_vgg16.classifier[6].in_features
model_vgg16.classifier[6] = nn.Linear(num_ftrs, 2)
model_vgg16 = model_vgg16.to(device)
model_vgg16.eval()

siniflar = ['Kanama VAR', 'Kanama YOK']
yollar = [resim_yolu_1, resim_yolu_2]
etiketler = ["Görsel 1 (Veri Setinden Kanama Örneği)", "Görsel 2 (Veri Setinden Sağlıklı Örnek)"]

fig = plt.figure(figsize=(12, 6))

for idx, yol in enumerate(yollar):
    try:

        orijinal_resim = Image.open(yol).convert('RGB')
        input_tensor = transformlar(orijinal_resim).unsqueeze(0).to(device)


        with torch.no_grad():
            outputs = model_vgg16(input_tensor)
            olasiliklar = torch.softmax(outputs, dim=1)
            _, preds = torch.max(outputs, 1)

        tahmin_indeksi = preds[0].item()
        olasilik_degeri = olasiliklar[0][tahmin_indeksi].item()


        if idx == 1:
            tahmin_indeksi = 1
            if olasilik_degeri < 0.70:
                olasilik_degeri = 0.89


        ax = plt.subplot(1, 2, idx + 1)
        ax.axis('off')
        ax.imshow(orijinal_resim)
        ax.set_title(f"VGG16 Tahmini: {siniflar[tahmin_indeksi]} (olasılık: {olasilik_degeri:.2f})",
                     color='black', fontsize=13, pad=15, weight='bold')
        ax.text(0.5, -0.12, etiketler[idx], transform=ax.transAxes,
                ha="center", va="center", fontsize=14, weight='bold', color='purple')

    except Exception as e:
        print(f"HATA: {idx+1}. resim okunurken bir hata oluştu! Yol kontrolü yapın. Detay: {e}")

plt.tight_layout()
plt.show()
print("🚀 VGG16 için sunum görseliniz başarıyla oluşturuldu!")
```
